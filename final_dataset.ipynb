```python
import pandas as pd


# scanf 함수에서 입력값의 길이를 제한하는 패턴 검출 함수
def check_scanf_vulnerability(row):
    if 'scanf(' in row:
        if '%[' in row and ']' in row:
            return 0  # %[버퍼크기]s 형태가 있으면 취약하지 않은 부분으로 판단
        else:
            return 1  # 그 외의 경우는 취약한 부분으로 판단
    return 0  # scanf 함수가 아닌 경우에는 취약하지 않은 부분으로 판단


# printf 함수에서 포맷스트링 처리를 하는 패턴 검출 함수
def check_printf_vulnerability(row):
    if 'printf(' in row:
        if '"' in row:
            input_index = row.index('printf(') + len('printf(')
            input_str = row[input_index:]
            if '%' not in input_str:
                return 1  # 포맷스트링이 없으면 취약한 부분으로 판단
            else:
                return 0  # 포맷스트링이 있으면 취약하지 않은 부분으로 판단
    return 0  # printf 함수가 아닌 경우에는 취약하지 않은 부분으로 판단

# strcpy 함수에서 버퍼 오버플로우 취약점을 검출하는 패턴 검출 함수
def check_strcpy_vulnerability(row):
    if 'strcpy(' in row:
        if 'sizeof(' not in row or ')' not in row:  # sizeof() 함수로 목적지 버퍼의 크기를 확인하는 코드가 없는 경우
            return 1  # 취약한 부분으로 판단
    return 0  # strcpy 함수가 사용되지 않았거나 버퍼의 크기를 확인하는 코드가 있는 경우에는 취약하지 않은 부분으로 판단



# 취약한 부분을 찾아내어 새로운 피처로 추가
train['vulnerable_strcpy'] = train['CODE'].apply(check_strcpy_vulnerability)
train['vulnerable_scanf'] = train['CODE'].apply(check_scanf_vulnerability)
train['vulnerable_printf'] = train['CODE'].apply(check_printf_vulnerability)

# 결과 확인
print(train.head())

```

```python
import re

# 동적 할당과 버퍼 사용 패턴을 식별하는 함수
def check_memory_allocation_and_buffer_use(code):
    # 동적 메모리 할당 패턴
    memory_alloc_pattern = re.compile(r'\b(malloc|calloc|realloc)\(([^)]+)\)')
    # 버퍼 사용 패턴
    buffer_use_pattern = re.compile(r'\b(strcpy|strcat|memcpy|sprintf)\(([^,]+),')

    # 메모리 할당 검색
    alloc_matches = memory_alloc_pattern.findall(code)
    buffer_use_matches = buffer_use_pattern.findall(code)

    # 할당된 메모리 사이즈와 버퍼 사용 인자 비교
    for alloc in alloc_matches:
        alloc_size = alloc[1]  # 할당 사이즈 인자 추출
        for buffer_use in buffer_use_matches:
            buffer_arg = buffer_use[1]  # 버퍼 사용 인자 추출
            if buffer_arg >= alloc_size:
                return 1
    return 0

# 새로운 컬럼에 동적 할당 메모리와 버퍼 사용 비교 결과 저장
train['buffer_allocation_check'] = train['CODE'].apply(check_memory_allocation_and_buffer_use)

# buffer_allocation_check 컬럼이 1인 행만 필터링하여 출력
#filtered_train = train[train['buffer_allocation_check'] == 1]
#print(filtered_train[['CODE', 'buffer_allocation_check']])
#train['buffer_allocation_check']
train
```

```python
# malloc, calloc, realloc, free 개수
def count_memory_operations(code):
    # 메모리 할당 및 해제 패턴
    malloc_pattern = re.compile(r'\bmalloc\(')
    calloc_pattern = re.compile(r'\bcalloc\(')
    realloc_pattern = re.compile(r'\brealloc\(')
    free_pattern = re.compile(r'\bfree\(')

    # 각 함수 호출 횟수 계산
    malloc_count = len(malloc_pattern.findall(code))
    calloc_count = len(calloc_pattern.findall(code))
    realloc_count = len(realloc_pattern.findall(code))
    free_count = len(free_pattern.findall(code))

    return malloc_count, calloc_count, realloc_count, free_count

# 데이터프레임에 새 컬럼 추가
train['malloc_count'], train['calloc_count'], train['realloc_count'], train['free_count'] = zip(*train['CODE'].apply(count_memory_operations))
```

```python
#각각의 할당 함수가 해제 되었는지 여부
def memory_management_status(code):
    malloc_pattern = re.compile(r'\bmalloc\(')
    calloc_pattern = re.compile(r'\bcalloc\(')
    realloc_pattern = re.compile(r'\brealloc\(')
    free_pattern = re.compile(r'\bfree\(')

    malloc_count = len(malloc_pattern.findall(code))
    calloc_count = len(calloc_pattern.findall(code))
    realloc_count = len(realloc_pattern.findall(code))
    free_count = len(free_pattern.findall(code))

    # 메모리 관리 상태 결정: 할당 횟수와 해제 횟수 비교
    malloc_status = 0 if malloc_count - free_count <= 0 else 1
    calloc_status = 0 if calloc_count - free_count <= 0 else 1
    realloc_status = 0 if realloc_count - free_count <= 0 else 1

    return malloc_status, calloc_status, realloc_status

# 데이터프레임에 새 컬럼 추가
train['malloc_status'], train['calloc_status'], train['realloc_status'] = zip(*train['CODE'].apply(memory_management_status))
```

```python
import pandas as pd
import re

# fopen(), fwrite(), fread(), malloc(), free() 함수 유무 확인
def check_file_memory_functions(code):
    file_memory_functions = ['fopen', 'fwrite', 'fread', 'malloc', 'free']
    for func in file_memory_functions:
        if re.search(r'\b' + func + r'\s*\(', code):
            return 1
    return 0

# pthread_t 전역 변수 유무 확인
def check_pthread_t_global_variable(code):
    if re.search(r'\bpthread_t\s+\w+', code):
        return 1
    return 0

# 공유 변수에 접근하기 전에 mutex로 동기화 여부 확인
def check_mutex_synchronization(code):
    if re.search(r'\bmutex\s*\.\s*acquire\s*\(', code):
        return 1
    return 0

# 각 피쳐를 계산하여 DataFrame에 추가
train['File_Memory_Functions_Exist'] = train['CODE'].apply(check_file_memory_functions)
train['pthread_t_Global_Variable_Exist'] = train['CODE'].apply(check_pthread_t_global_variable)
train['Mutex_Synchronization'] = train['CODE'].apply(check_mutex_synchronization)

# 결과 확인
train
```

```python
import pandas as pd
import re

# 포인터 초기화
def pointer_initialization(code):
    if re.search(r'\bchar\s*\*\s*\w+\s*=\s*NULL\s*;', code):
        return 1
    else:
        return 0

# NULL 체크
def null_check(code):
    if re.search(r'\bif\s*\(\s*\w+\s*!=\s*NULL\s*\)\s*{', code):
        return 1
    else:
        return 0

# 동적 메모리 할당 후 검사
def dynamic_memory_allocation_and_check(code):
    if re.search(r'\b\w+\s*\*\s*\w+\s*=\s*malloc\s*\(\s*\w*\s*\)\s*;\s*if\s*\(\s*\w+\s*==\s*NULL\s*\)\s*{', code):
        return 1
    else:
        return 0

# 포인터 사용 후 NULL로 초기화
def pointer_reset_after_use(code):
    if re.search(r'\bfree\s*\(\s*\w+\s*\)\s*;\s*\w+\s*=\s*NULL\s*;', code):
        return 1
    else:
        return 0

# 각 피쳐를 계산하여 DataFrame에 추가
train['Pointer_Initialization'] = train['CODE'].apply(pointer_initialization)
train['NULL_Check'] = train['CODE'].apply(null_check)
train['Dynamic_Memory_Allocation_And_Check'] = train['CODE'].apply(dynamic_memory_allocation_and_check)
train['Pointer_Reset_After_Use'] = train['CODE'].apply(pointer_reset_after_use)

# 결과 확인
train
```

```python
import pandas as pd
import re

# free() 유무 검사
def check_for_free(code):
    if re.search(r'\bfree\s*\(\s*\w+\s*\)\s*;\s*\w+\s*=\s*NULL\s*;', code):
        return 1
    else:
        return 0

# 포인터 역참조 유무
def check_for_pointer_dereference(code):
    if re.search(r'\b\*\w+\s*=', code):
        return 1
    else:
        return 0

# 배열의 인덱스
def array_index(code):
    if re.search(r'\bchar\s+\w+\[\s*0x[0-9a-fA-F]+\s*\];', code):
        return 1
    else:
        return 0

# 각 피쳐를 계산하여 DataFrame에 추가
train['Check_for_Free'] = train['CODE'].apply(check_for_free)
train['Check_for_Pointer_Dereference'] = train['CODE'].apply(check_for_pointer_dereference)
train['Array_Index'] = train['CODE'].apply(array_index)

# 결과 확인
train
```

```python
import re

# 배열의 인덱스를 확인하여 CWE-125 취약점을 분류하는 함수
def check_array_index_vulnerability(row):
    # 변수 이름을 찾기 위한 정규 표현식 패턴
    pattern = r'\b(\w+)\[(\w+)\]'

    # 정규 표현식을 사용하여 변수 이름과 인덱스를 추출
    matches = re.findall(pattern, row)
    for match in matches:
        var_name, index = match

        # 변수 이름이 포함된 부분을 제거한 후, 변수 할당식을 찾기 위한 패턴
        assignment_pattern = re.compile(r'\b' + re.escape(var_name) + r'\s*=\s*(.*)')
        assignment_match = assignment_pattern.search(row)
        if assignment_match:
            buffer_size = assignment_match.group(1)

            # 인덱스와 버퍼 크기를 비교하여 취약점을 판단
            if index.isdigit() and buffer_size.isdigit():
                if int(index) >= int(buffer_size):
                    return 1
                else:
                    return 0
    return 0
# 취약한 부분을 찾아내어 새로운 피처로 추가
train['array_index_vulnerability'] = train['CODE'].apply(check_array_index_vulnerability)

# 결과 확인
train
```

```python
import pandas as pd
import re

def count_data_type_usage(code, data_type):
    pattern = r'\b' + re.escape(data_type) + r'\b'
    return len(re.findall(pattern, code))

# 모든 정수 자료형 목록
data_types = [
    'int', 'short', 'long', 'long long',
    'unsigned int', 'unsigned short', 'unsigned long', 'unsigned long long'
]

# 각 자료형에 대한 사용 횟수를 컬럼으로 추가
for data_type in data_types:
    train[data_type] = train['CODE'].apply(lambda code: count_data_type_usage(code, data_type))

# 결과 확인
train.head()
```

```python
# 코드의 줄 수를 계산하는 함수
def count_lines(code):
    return len(code.split('\n'))

# 'line_count' 컬럼에 각 코드의 줄 수 저장
train['line_count'] = train['CODE'].apply(count_lines)

# 결과 확인
train
```
