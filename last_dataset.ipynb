```python
import pandas as pd

# 파일 경로 리스트
file_paths = [
    '/content/drive/MyDrive/Colab Notebooks/plus data/DiverseVul_3_plus(complete).csv',
    '/content/drive/MyDrive/Colab Notebooks/Julite_data/Labels_Safe_MSR+Juliet_random.csv'

]
# DataFrame들을 담을 리스트
dfs = []

# 각 파일을 읽어들여 리스트에 추가
for file_path in file_paths:
    dfs.append(pd.read_csv(file_path))

# DataFrame들을 하나로 합치기
combined_df = pd.concat(dfs, ignore_index=True)

# 하나의 CSV 파일로 저장
combined_df.to_csv('/content/drive/MyDrive/Colab Notebooks/model train/Last.csv', index=False)
```
