```python

# 'vul' 컬럼 생성 및 조건에 따라 값 할당
train['vul'] = train.apply(lambda row: 1 if pd.notna(row['code_before']) else 0, axis=1)

# 결과 출력
train
```

```python
# 'func_before'를 'CODE'로 이름 변경
train.rename(columns={'code_before': 'CODE'}, inplace=True)

# 'func_after'를 새로운 행으로 추가하여 'CODE' 컬럼에 병합
train_after = train.copy()
train_after['CODE'] = train_after['code_after']

# 'func_after' 컬럼을 'CODE' 컬럼에 추가한 후 'func_before'와 중복된 데이터프레임 제거
train_combined = pd.concat([train, train_after], ignore_index=True)

# 'func_after' 컬럼 삭제
train_combined.drop(columns=['code_after'], inplace=True)

# 데이터셋 확인
train = train_combined
```

```python
import pandas as pd
import numpy as np  # NaN 값을 다루기 위해 numpy 라이브러리를 사용합니다.

# CSV 파일 경로
file_path = "/content/drive/MyDrive/Colab Notebooks/web_dataset.csv"

# CSV 파일 읽어오기
df = pd.read_csv(file_path)

# XSS 취약점 패턴 확인 함수
def detect_xss_vulnerability(code):
    if isinstance(code, float) and np.isnan(code):  # NaN 값을 처리합니다.
        return 0  # NaN 값이면 취약점이 없음을 반환합니다.

    xss_patterns = ["echo", "print", "innerHTML", "document.write", "<%=", "%>", "getParameter",
                    "eval", "alert", "script", "onload", "onerror", "javascript:"]  # 추가 패턴
    for pattern in xss_patterns:
        if pattern in code:
            return 1  # 취약점 발견
    return 0  # 취약점 없음

# XSS 취약점 여부 컬럼 추가
df["XSS_Vulnerability"] = df["CODE"].apply(detect_xss_vulnerability)

# 결과 확인
print(df)
```

```python
import pandas as pd
import numpy as np  # NaN 값을 다루기 위해 numpy 라이브러리를 사용합니다.

# CWE-89 SQL Injection 취약점 패턴 확인 함수
def detect_sql_injection_vulnerability(code):
    if isinstance(code, float) and np.isnan(code):  # NaN 값을 처리합니다.
        return 0  # NaN 값이면 취약점이 없음을 반환합니다.

    sql_patterns = ["mysqli_query", "PDO::query", "connection.prepare", "Statement.executeQuery",
                    "createNamedQuery"]  # SQL Injection 취약점과 관련된 패턴
    for pattern in sql_patterns:
        if pattern in code:
            return 1  # 취약점 발견
    return 0  # 취약점 없음

# SQL Injection 취약점 여부 컬럼 추가
df["SQL_Injection_Vulnerability"] = df["CODE"].apply(detect_sql_injection_vulnerability)

# 결과 확인
print(df)
```

```python
import pandas as pd

# CSRF 취약점 패턴 확인 함수
def detect_csrf_vulnerability(code):
    if isinstance(code, float) and pd.isnull(code):  # NaN 값을 처리합니다.
        return 0  # NaN 값이면 취약점이 없음을 반환합니다.

    csrf_patterns = ["CSRF", "Cross-Site Request Forgery", "csrf_token", "CSRF_TOKEN", "csrfToken"]  # CSRF 취약점과 관련된 패턴
    for pattern in csrf_patterns:
        if pattern in code:
            return 1  # 취약점 발견
    return 0  # 취약점 없음

# CSRF 취약점 여부 컬럼 추가
df["CSRF_Vulnerability"] = df["CODE"].apply(detect_csrf_vulnerability)

# 결과 확인
print(df)
```

```python
import pandas as pd
import re

# Path Traversal 취약점 패턴 확인 함수
def detect_path_traversal_vulnerability(code):
    if isinstance(code, float) and pd.isnull(code):  # NaN 값을 처리합니다.
        return 0  # NaN 값이면 취약점이 없음을 반환합니다.

    path_traversal_patterns = ["../../", "../", "..\\", "..%2f", "..%5c"]  # Path Traversal 취약점과 관련된 패턴
    for pattern in path_traversal_patterns:
        if pattern in code:
            return 1  # 취약점 발견

    return 0  # 취약점 없음

# Path Traversal 취약점 여부 컬럼 추가
df["Path_Traversal_Vulnerability"] = df["CODE"].apply(detect_path_traversal_vulnerability)

# 결과 확인
print(df)
```

```python
import pandas as pd

# SSRF 취약점 패턴 확인 함수
def detect_ssrf_vulnerability(code):
    if isinstance(code, float) and pd.isnull(code):  # NaN 값을 처리합니다.
        return 0  # NaN 값이면 취약점이 없음을 반환합니다.

    ssrf_functions = ["curl_exec", "file_get_contents", "http.get", "axios.get", "fetch", "URL.openConnection", "HttpClient.execute"]
    for func in ssrf_functions:
        if func in code:
            return 1  # 취약점 발견

    return 0  # 취약점 없음

# SSRF 취약점 여부 컬럼 추가
df["SSRF_Vulnerability"] = df["CODE"].apply(detect_ssrf_vulnerability)

# 결과 확인
print(df)
```

```python
import pandas as pd
import re

# 외부 URL 요청을 확인하여 SSRF 취약점 패턴 탐지하는 함수
def detect_ssrf_vulnerability_alternate(func_code):
    if isinstance(func_code, float) and pd.isnull(func_code):  # NaN 값을 처리합니다.
        return 0  # NaN 값이면 취약점이 없음을 반환합니다.

    # 허용되는 도메인 또는 IP 주소 목록
    allowed_domains = ["example.com", "192.168.0.1"]

    # 입력된 코드에서 외부 URL을 추출합니다.
    urls = re.findall(r'https?://(?:[a-fA-F0-9.:]+)', func_code)
    if urls:
        for url in urls:
            # URL을 통해 호스트 부분을 추출합니다.
            host = re.search(r'https?://(?:[a-fA-F0-9.:]+)', url).group(0)

            # 허용되지 않는 도메인 또는 IP 주소인 경우 취약점으로 판단
            if host not in allowed_domains:
                return 1  # 취약점 발견

    return 0  # 취약점 없음

# SSRF 취약점 여부를 나타내는 새로운 컬럼 추가
df["SSRF_Vuln_Alternate"] = df["CODE"].apply(detect_ssrf_vulnerability_alternate)

# 결과 확인
print(df)
```

```python
import pandas as pd
import re

# 함수 및 패턴이 코드에 나오는지 확인하고 횟수를 반환하는 함수
def count_functions_or_patterns(func_string, row):
    # 데이터를 문자열로 변환하여 처리합니다.
    row = str(row)
    return row.count(func_string)

# 각 언어에 대해 자주 사용되는 함수 및 패턴들
php_functions_to_check = ['curl_exec(', 'file_get_contents(', 'echo ', 'print ', 'htmlspecialchars(', 'strip_tags(']
javascript_functions_to_check = ['http.get(', 'axios.get(', 'fetch(', 'innerHTML ', 'document.write(', 'document.getElementById(']
java_functions_to_check = ['URL.openConnection()', 'HttpClient.execute()', 'PreparedStatement', 'createNamedQuery(']
other_patterns_to_check = ['eval(', 'exec(', 'system(', 'shell_exec(', 'popen(', 'pcntl_exec(', 'assert(']  # 다른 언어에서 자주 사용되는 패턴들

# 추가할 다른 함수들 및 패턴들
php_other_functions_to_check = ['header(', 'setcookie(', 'mail(', 'preg_replace(']  # PHP에서 자주 사용되는 다른 함수들
javascript_other_functions_to_check = ['setTimeout(', 'setInterval(', 'XMLHttpRequest(', 'eval(']  # JavaScript에서 자주 사용되는 다른 함수들
java_other_functions_to_check = ['System.out.println(', 'System.err.println(', 'FileReader(', 'FileWriter(']  # Java에서 자주 사용되는 다른 함수들

# PHP 함수 및 패턴 확인 및 횟수를 새로운 피처로 추가
for func in php_functions_to_check + php_other_functions_to_check:
    train[f'PHP_{func}'] = train['CODE'].apply(lambda row: count_functions_or_patterns(func, row))

# JavaScript 함수 및 패턴 확인 및 횟수를 새로운 피처로 추가
for func in javascript_functions_to_check + javascript_other_functions_to_check:
    train[f'JavaScript_{func}'] = train['CODE'].apply(lambda row: count_functions_or_patterns(func, row))

# Java 함수 및 패턴 확인 및 횟수를 새로운 피처로 추가
for func in java_functions_to_check + java_other_functions_to_check:
    train[f'Java_{func}'] = train['CODE'].apply(lambda row: count_functions_or_patterns(func, row))

# 다른 패턴 확인 및 횟수를 새로운 피처로 추가
for pattern in other_patterns_to_check:
    train[f'Other_Pattern_{pattern}'] = train['CODE'].apply(lambda row: count_functions_or_patterns(pattern, row))

# 결과 확인
print(train)
```
