```python
import pandas as pd
import numpy as np  # NaN ê°’ì„ ë‹¤ë£¨ê¸° ìœ„í•´ numpy ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.

# CSV íŒŒì¼ ê²½ë¡œ
file_path = "/content/drive/MyDrive/Colab Notebooks/web_dataset.csv"

# CSV íŒŒì¼ ì½ì–´ì˜¤ê¸°
df = pd.read_csv(file_path)

# XSS ì·¨ì•½ì  íŒ¨í„´ í™•ì¸ í•¨ìˆ˜
def detect_xss_vulnerability(code):
    if isinstance(code, float) and np.isnan(code):  # NaN ê°’ì„ ì²˜ë¦¬í•©ë‹ˆë‹¤.
        return 0  # NaN ê°’ì´ë©´ ì·¨ì•½ì ì´ ì—†ìŒì„ ë°˜í™˜í•©ë‹ˆë‹¤.

    xss_patterns = ["echo", "print", "innerHTML", "document.write", "<%=", "%>", "getParameter",
                    "eval", "alert", "script", "onload", "onerror", "javascript:"]  # ì¶”ê°€ íŒ¨í„´
    for pattern in xss_patterns:
        if pattern in code:
            return 1  # ì·¨ì•½ì  ë°œê²¬
    return 0  # ì·¨ì•½ì  ì—†ìŒ

# XSS ì·¨ì•½ì  ì—¬ë¶€ ì»¬ëŸ¼ ì¶”ê°€
df["XSS_Vulnerability"] = df["CODE"].apply(detect_xss_vulnerability)

# ê²°ê³¼ í™•ì¸
print(df)
```

```python
import pandas as pd
import numpy as np  # NaN ê°’ì„ ë‹¤ë£¨ê¸° ìœ„í•´ numpy ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.

# CWE-89 SQL Injection ì·¨ì•½ì  íŒ¨í„´ í™•ì¸ í•¨ìˆ˜
def detect_sql_injection_vulnerability(code):
    if isinstance(code, float) and np.isnan(code):  # NaN ê°’ì„ ì²˜ë¦¬í•©ë‹ˆë‹¤.
        return 0  # NaN ê°’ì´ë©´ ì·¨ì•½ì ì´ ì—†ìŒì„ ë°˜í™˜í•©ë‹ˆë‹¤.

    sql_patterns = ["mysqli_query", "PDO::query", "connection.prepare", "Statement.executeQuery",
                    "createNamedQuery"]  # SQL Injection ì·¨ì•½ì ê³¼ ê´€ë ¨ëœ íŒ¨í„´
    for pattern in sql_patterns:
        if pattern in code:
            return 1  # ì·¨ì•½ì  ë°œê²¬
    return 0  # ì·¨ì•½ì  ì—†ìŒ

# SQL Injection ì·¨ì•½ì  ì—¬ë¶€ ì»¬ëŸ¼ ì¶”ê°€
df["SQL_Injection_Vulnerability"] = df["CODE"].apply(detect_sql_injection_vulnerability)

# ê²°ê³¼ í™•ì¸
print(df)
```

```python
import pandas as pd

# CSRF ì·¨ì•½ì  íŒ¨í„´ í™•ì¸ í•¨ìˆ˜
def detect_csrf_vulnerability(code):
    if isinstance(code, float) and pd.isnull(code):  # NaN ê°’ì„ ì²˜ë¦¬í•©ë‹ˆë‹¤.
        return 0  # NaN ê°’ì´ë©´ ì·¨ì•½ì ì´ ì—†ìŒì„ ë°˜í™˜í•©ë‹ˆë‹¤.

    csrf_patterns = ["CSRF", "Cross-Site Request Forgery", "csrf_token", "CSRF_TOKEN", "csrfToken"]  # CSRF ì·¨ì•½ì ê³¼ ê´€ë ¨ëœ íŒ¨í„´
    for pattern in csrf_patterns:
        if pattern in code:
            return 1  # ì·¨ì•½ì  ë°œê²¬
    return 0  # ì·¨ì•½ì  ì—†ìŒ

# CSRF ì·¨ì•½ì  ì—¬ë¶€ ì»¬ëŸ¼ ì¶”ê°€
df["CSRF_Vulnerability"] = df["CODE"].apply(detect_csrf_vulnerability)

# ê²°ê³¼ í™•ì¸
print(df)
```

```python
import pandas as pd
import re

# Path Traversal ì·¨ì•½ì  íŒ¨í„´ í™•ì¸ í•¨ìˆ˜
def detect_path_traversal_vulnerability(code):
    if isinstance(code, float) and pd.isnull(code):  # NaN ê°’ì„ ì²˜ë¦¬í•©ë‹ˆë‹¤.
        return 0  # NaN ê°’ì´ë©´ ì·¨ì•½ì ì´ ì—†ìŒì„ ë°˜í™˜í•©ë‹ˆë‹¤.

    path_traversal_patterns = ["../../", "../", "..\\", "..%2f", "..%5c"]  # Path Traversal ì·¨ì•½ì ê³¼ ê´€ë ¨ëœ íŒ¨í„´
    for pattern in path_traversal_patterns:
        if pattern in code:
            return 1  # ì·¨ì•½ì  ë°œê²¬

    return 0  # ì·¨ì•½ì  ì—†ìŒ

# Path Traversal ì·¨ì•½ì  ì—¬ë¶€ ì»¬ëŸ¼ ì¶”ê°€
df["Path_Traversal_Vulnerability"] = df["CODE"].apply(detect_path_traversal_vulnerability)

# ê²°ê³¼ í™•ì¸
print(df)
```

```python
import pandas as pd

# SSRF ì·¨ì•½ì  íŒ¨í„´ í™•ì¸ í•¨ìˆ˜
def detect_ssrf_vulnerability(code):
    if isinstance(code, float) and pd.isnull(code):  # NaN ê°’ì„ ì²˜ë¦¬í•©ë‹ˆë‹¤.
        return 0  # NaN ê°’ì´ë©´ ì·¨ì•½ì ì´ ì—†ìŒì„ ë°˜í™˜í•©ë‹ˆë‹¤.

    ssrf_functions = ["curl_exec", "file_get_contents", "http.get", "axios.get", "fetch", "URL.openConnection", "HttpClient.execute"]
    for func in ssrf_functions:
        if func in code:
            return 1  # ì·¨ì•½ì  ë°œê²¬

    return 0  # ì·¨ì•½ì  ì—†ìŒ

# SSRF ì·¨ì•½ì  ì—¬ë¶€ ì»¬ëŸ¼ ì¶”ê°€
df["SSRF_Vulnerability"] = df["CODE"].apply(detect_ssrf_vulnerability)

# ê²°ê³¼ í™•ì¸
print(df)
```

```python
import pandas as pd
import re

# ì™¸ë¶€ URL ìš”ì²­ì„ í™•ì¸í•˜ì—¬ SSRF ì·¨ì•½ì  íŒ¨í„´ íƒì§€í•˜ëŠ” í•¨ìˆ˜
def detect_ssrf_vulnerability_alternate(func_code):
    if isinstance(func_code, float) and pd.isnull(func_code):  # NaN ê°’ì„ ì²˜ë¦¬í•©ë‹ˆë‹¤.
        return 0  # NaN ê°’ì´ë©´ ì·¨ì•½ì ì´ ì—†ìŒì„ ë°˜í™˜í•©ë‹ˆë‹¤.

    # í—ˆìš©ë˜ëŠ” ë„ë©”ì¸ ë˜ëŠ” IP ì£¼ì†Œ ëª©ë¡
    allowed_domains = ["example.com", "192.168.0.1"]

    # ìž…ë ¥ëœ ì½”ë“œì—ì„œ ì™¸ë¶€ URLì„ ì¶”ì¶œí•©ë‹ˆë‹¤.
    urls = re.findall(r'https?://(?:[a-fA-F0-9.:]+)', func_code)
    if urls:
        for url in urls:
            # URLì„ í†µí•´ í˜¸ìŠ¤íŠ¸ ë¶€ë¶„ì„ ì¶”ì¶œí•©ë‹ˆë‹¤.
            host = re.search(r'https?://(?:[a-fA-F0-9.:]+)', url).group(0)

            # í—ˆìš©ë˜ì§€ ì•ŠëŠ” ë„ë©”ì¸ ë˜ëŠ” IP ì£¼ì†Œì¸ ê²½ìš° ì·¨ì•½ì ìœ¼ë¡œ íŒë‹¨
            if host not in allowed_domains:
                return 1  # ì·¨ì•½ì  ë°œê²¬

    return 0  # ì·¨ì•½ì  ì—†ìŒ

# SSRF ì·¨ì•½ì  ì—¬ë¶€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ìƒˆë¡œìš´ ì»¬ëŸ¼ ì¶”ê°€
df["SSRF_Vuln_Alternate"] = df["CODE"].apply(detect_ssrf_vulnerability_alternate)

# ê²°ê³¼ í™•ì¸
print(df)
```

```python
import pandas as pd
import re

# í•¨ìˆ˜ ë° íŒ¨í„´ì´ ì½”ë“œì— ë‚˜ì˜¤ëŠ”ì§€ í™•ì¸í•˜ê³  íšŸìˆ˜ë¥¼ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜
def count_functions_or_patterns(func_string, row):
    # ë°ì´í„°ë¥¼ ë¬¸ìžì—´ë¡œ ë³€í™˜í•˜ì—¬ ì²˜ë¦¬í•©ë‹ˆë‹¤.
    row = str(row)
    return row.count(func_string)

# ê° ì–¸ì–´ì— ëŒ€í•´ ìžì£¼ ì‚¬ìš©ë˜ëŠ” í•¨ìˆ˜ ë° íŒ¨í„´ë“¤
php_functions_to_check = ['curl_exec(', 'file_get_contents(', 'echo ', 'print ', 'htmlspecialchars(', 'strip_tags(']
javascript_functions_to_check = ['http.get(', 'axios.get(', 'fetch(', 'innerHTML ', 'document.write(', 'document.getElementById(']
java_functions_to_check = ['URL.openConnection()', 'HttpClient.execute()', 'PreparedStatement', 'createNamedQuery(']
other_patterns_to_check = ['eval(', 'exec(', 'system(', 'shell_exec(', 'popen(', 'pcntl_exec(', 'assert(']  # ë‹¤ë¥¸ ì–¸ì–´ì—ì„œ ìžì£¼ ì‚¬ìš©ë˜ëŠ” íŒ¨í„´ë“¤

# ì¶”ê°€í•  ë‹¤ë¥¸ í•¨ìˆ˜ë“¤ ë° íŒ¨í„´ë“¤
php_other_functions_to_check = ['header(', 'setcookie(', 'mail(', 'preg_replace(']  # PHPì—ì„œ ìžì£¼ ì‚¬ìš©ë˜ëŠ” ë‹¤ë¥¸ í•¨ìˆ˜ë“¤
javascript_other_functions_to_check = ['setTimeout(', 'setInterval(', 'XMLHttpRequest(', 'eval(']  # JavaScriptì—ì„œ ìžì£¼ ì‚¬ìš©ë˜ëŠ” ë‹¤ë¥¸ í•¨ìˆ˜ë“¤
java_other_functions_to_check = ['System.out.println(', 'System.err.println(', 'FileReader(', 'FileWriter(']  # Javaì—ì„œ ìžì£¼ ì‚¬ìš©ë˜ëŠ” ë‹¤ë¥¸ í•¨ìˆ˜ë“¤

# PHP í•¨ìˆ˜ ë° íŒ¨í„´ í™•ì¸ ë° íšŸìˆ˜ë¥¼ ìƒˆë¡œìš´ í”¼ì²˜ë¡œ ì¶”ê°€
for func in php_functions_to_check + php_other_functions_to_check:
    train[f'PHP_{func}'] = train['CODE'].apply(lambda row: count_functions_or_patterns(func, row))

# JavaScript í•¨ìˆ˜ ë° íŒ¨í„´ í™•ì¸ ë° íšŸìˆ˜ë¥¼ ìƒˆë¡œìš´ í”¼ì²˜ë¡œ ì¶”ê°€
for func in javascript_functions_to_check + javascript_other_functions_to_check:
    train[f'JavaScript_{func}'] = train['CODE'].apply(lambda row: count_functions_or_patterns(func, row))

# Java í•¨ìˆ˜ ë° íŒ¨í„´ í™•ì¸ ë° íšŸìˆ˜ë¥¼ ìƒˆë¡œìš´ í”¼ì²˜ë¡œ ì¶”ê°€
for func in java_functions_to_check + java_other_functions_to_check:
    train[f'Java_{func}'] = train['CODE'].apply(lambda row: count_functions_or_patterns(func, row))

# ë‹¤ë¥¸ íŒ¨í„´ í™•ì¸ ë° íšŸìˆ˜ë¥¼ ìƒˆë¡œìš´ í”¼ì²˜ë¡œ ì¶”ê°€
for pattern in other_patterns_to_check:
    train[f'Other_Pattern_{pattern}'] = train['CODE'].apply(lambda row: count_functions_or_patterns(pattern, row))

# ê²°ê³¼ í™•ì¸
print(train)
```
