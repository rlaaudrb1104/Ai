```python
import pandas as pd
train = pd.read_csv('/content/drive/MyDrive/Data/MSR_data_cleaned.csv')
train.columns
```

```python
selected_columns = ['CWE ID', 'func_after', 'func_before', 'lang', 'vul']
df = train[selected_columns]
df
```

```python
filtered_df = df[df['vul'] == 1]
filtered_df
```

```python
# 'code' 컬럼 생성
df['code'] = ''

# 'vul'이 0일 때, 'func_before'를 'code'에 저장
df.loc[df['vul'] == 0, 'code'] = df['func_before']

# 'vul'이 1일 때, 'func_before'를 'code'에 저장
df.loc[df['vul'] == 1, 'code'] = df['func_before']

# 'vul'이 1인 경우, 'func_after'을 'vul'을 0으로 하여 'code'로 추가
df_after = df[df['vul'] == 1].copy()
df_after['vul'] = 0
df_after['code'] = df_after['func_after']

# 'func_before'와 'func_after' 컬럼 삭제
df.drop(['func_before', 'func_after'], axis=1, inplace=True)
df_after.drop(['func_before', 'func_after'], axis=1, inplace=True)
```

```python
filtered_df = df[df['vul'] == 1]
filtered_df
```
