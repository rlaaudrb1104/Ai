```python
import re
def preprocess(df, file_name):
    # 멀티 라인 주석 제거
    df['code'] = df['code'].replace(re.compile(r'/\*.*?\*/', re.DOTALL), "", regex=True)
    df.to_csv(file_name, index=False)
    return df  # 수정된 데이터프레임 반환
```

```python
preprocess(df,"/content/drive/MyDrive/Dataset/final/CWE415_preprocess.csv")
```

```python
pattern = r'\b\w*(CWE|bad|good)\w*\b'

df3['code'] = df3['code'].apply(lambda x: re.sub(pattern, 'func', x, flags=re.IGNORECASE))

df3
df3.to_csv("/content/drive/MyDrive/Dataset/final/CWE415_final_preprocess.csv", index=False)
```

```python
def preprocess2(df, file_name):
    # 멀티 라인 주석 제거 및 함수 이름 변경
    # good, bad, CWE로 시작하는 모든 함수 이름을 'func()'으로 변경
    df['code'] = df['code'].replace(re.compile(r'\b\w*(CWE|bad|good)\w*\b'), 'func', regex=True)
    df.to_csv(file_name, index=False)
    return df  # 수정된 데이터프레임 반환
```
