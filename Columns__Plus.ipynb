```python
import pandas as pd

train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MSR_dataset.csv')
train
```

```python
import pandas as pd

# 데이터셋 로드
df = train

# 각 함수들을 확인하여 해당 함수가 나오는지와 횟수를 반환하는 함수
def count_functions(func_string, row):
    return row.count(func_string)

# 함수 목록
functions_to_check = ['assert', 'atoi', 'atof', 'calloc', 'errno', 'exit', 'fclose', 'fopen', 'free','gets','read','scanf','printf','write','strcpy','strcmp','memcpy','malloc','realloc','puts','strcat','sprintf','strtok','signal','NULL']

# 각 함수들이 나오는지 확인하고 횟수를 새로운 피처로 추가
for func in functions_to_check:
    df[f'{func}_count'] = df['func_before'].apply(lambda row: count_functions(func, row))


```
