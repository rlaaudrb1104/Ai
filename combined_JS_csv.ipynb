```python
import json
import csv
from google.colab import drive

# Google Drive 마운트
drive.mount('/content/drive')

def json_to_csv(json_file, csv_file):
    try:
        with open(json_file, 'r', encoding='utf-8') as jfile:
            data = json.load(jfile)
    except json.decoder.JSONDecodeError as e:
        print(f"Error decoding JSON in file: {json_file}")
        print(f"Error message: {str(e)}")
        return

    # CSV 파일 쓰기
    with open(csv_file, 'w', newline='', encoding='utf-8') as cfile:
        writer = csv.writer(cfile)

        # CSV 파일의 헤더를 쓰기
        writer.writerow(data[0].keys())

        # JSON 데이터를 CSV 파일에 쓰기
        for item in data:
            writer.writerow(item.values())

# JSON 파일과 CSV 파일 경로 지정
json_file_path = '/content/drive/MyDrive/plus_data/data_JavaScript_vul.json'
csv_file_path = '/content/drive/MyDrive/plus_data/data_JavaScript_vul.csv'

# JSON 파일을 CSV 파일로 변환
json_to_csv(json_file_path, csv_file_path)
```
