```python
import pandas as pd

def clean_code(code):
    # 줄바꿈을 찾아서 세미콜론을 추가
    code = code.replace('\n', ';\n')
    # 중괄호 주변에 공백 추가
    code = code.replace('{', ' {\n\t')
    code = code.replace('}', '\n}')
    return code

def clean_dataset(input_file, output_file):
    # 데이터셋 불러오기
    dataset = pd.read_csv(input_file)
    # 코드 정리
    dataset['code'] = dataset['code'].apply(clean_code)
    # 정리된 데이터셋 저장
    dataset.to_csv(output_file, index=False)

# 데이터셋 파일 경로
input_file = '/content/drive/MyDrive/Colab Notebooks/Julite_data/julite_dataset.csv'
output_file = '/content/drive/MyDrive/Colab Notebooks/Julite_data/julite_dataset2.csv'

# 데이터셋 정리 함수 호출
clean_dataset(input_file, output_file)
```

```python
#MSR no_random
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Julite_data/MSR_dataset_norandom.csv')

df_counts = df[["CWE ID", 'vul']].value_counts()
print(df_counts)
```

```python
import pandas as pd

def cut_data(df, cwe_id, count):
    """
    데이터프레임(df)에서 특정 CWE ID(cwe_id)의 vul이 0이고, count만큼의 행을 잘라내는 함수
    """
    df_cwe = df[(df['CWE ID'] == cwe_id) & (df['vul'] == 0)].head(count)
    return df_cwe

# 데이터 불러오기
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Julite_data/MSR_dataset_norandom.csv')

# 잘라낼 각 CWE ID에 해당하는 행의 개수 지정
cwe_counts = {
    'CWE-20': 300,
    'CWE-119': 300,
    'CWE-399': 300,
    'CWE-125': 300,
    'CWE-415': 150,
    'CWE-416': 150,
    'CWE-190': 150
}

# 각 CWE ID에 해당하는 개수만큼 데이터 잘라내기
cut_dfs = []
for cwe_id, count in cwe_counts.items():
    cut_dfs.append(cut_data(df, cwe_id, count))

# vul이 1인 데이터는 따로 저장
df_vul_1 = df[df['vul'] == 1]

# 잘라낸 데이터프레임과 vul이 1인 데이터를 합치기
df_cut = pd.concat(cut_dfs + [df_vul_1])
df_cut.reset_index(drop=True, inplace=True)
print(df_cut)
```

```python
df_counts = df_cut[["CWE ID", 'vul']].value_counts()
print(df_counts)
```
