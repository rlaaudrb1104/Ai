```python
import os
import pandas as pd

# 원본 데이터셋 로드
file_path = "/content/drive/MyDrive/Data/final_train.csv"
df_original = pd.read_csv(file_path)

# CWE ID 컬럼이 'CWE-119'인 행만 골라내기
cwe_119_df = df_original[df_original['CWE ID'] == 'CWE-119']

# vul이 1인 행이 2000개만 남기고 나머지 삭제
cwe_119_vul_1 = cwe_119_df[cwe_119_df['vul'] == 1].sample(n=2000, random_state=42)

# vul이 0인 행이 1000개만 남기고 나머지 삭제
cwe_119_vul_0 = cwe_119_df[cwe_119_df['vul'] == 0].sample(n=1000, random_state=42)

# 처리된 데이터를 합치기
final_df = pd.concat([cwe_119_vul_1, cwe_119_vul_0])

# 원본 데이터셋에서 CWE-119 관련 행 제거하기
df_original = df_original[df_original['CWE ID'] != 'CWE-119']

# 합쳐진 데이터셋 만들기
merged_df = pd.concat([df_original, final_df])

# 인덱스 초기화
merged_df.reset_index(drop=True, inplace=True)

# 결과를 확인
merged_df.to_csv('/content/drive/MyDrive/Data/final_train.csv',index=False)
```

```python
import os
import pandas as pd

# 원본 데이터셋 로드
file_path = "/content/drive/MyDrive/Data/final_train.csv"
df_original = pd.read_csv(file_path)

# CWE ID 컬럼이 'CWE-20'이고 vul이 0인 행이 1000개만 남기고 나머지 삭제
cwe_20_vul_0 = df_original[(df_original['CWE ID'] == 'CWE-20') & (df_original['vul'] == 0)].sample(n=1000, random_state=42)

# 원본 데이터셋에서 해당 행 제거하기
df_original = df_original[~((df_original['CWE ID'] == 'CWE-20') & (df_original['vul'] == 0))]

# 합쳐진 데이터셋 만들기
merged_df = pd.concat([df_original, cwe_20_vul_0])

# 인덱스 초기화
merged_df.reset_index(drop=True, inplace=True)

# 결과를 확인
merged_df
merged_df.to_csv('/content/drive/MyDrive/Data/final_train.csv',index=False)
```

```python
import os
import pandas as pd

# 원본 데이터셋 로드
file_path = "/content/drive/MyDrive/Data/final_train.csv"
df_original = pd.read_csv(file_path)

# 남겨야 할 데이터의 개수
keep_count = 800

# CWE ID 컬럼이 지정된 값이면서 vul이 0인 데이터 중에서 남겨야 할 개수만큼만 샘플링하여 추출
cwe_list = ['CWE-78', 'CWE-122', 'CWE-121', 'CWE-415', 'CWE-399', 'CWE-190', 'CWE-125', 'CWE-416']
filtered_data = pd.DataFrame()
for cwe_id in cwe_list:
    temp_df = df_original[(df_original['CWE ID'] == cwe_id) & (df_original['vul'] == 0)].sample(n=keep_count, random_state=42)
    filtered_data = pd.concat([filtered_data, temp_df])

# 추출한 데이터를 제외한 나머지 데이터 제거
df_original = df_original[~((df_original['CWE ID'].isin(cwe_list)) & (df_original['vul'] == 0))]

# 추출한 데이터와 기존 데이터 합치기
merged_df = pd.concat([df_original, filtered_data])

# 인덱스 초기화
merged_df.reset_index(drop=True, inplace=True)

# 결과를 확인
merged_df
merged_df.to_csv('/content/drive/MyDrive/Data/final_train.csv',index=False)
```

```python
import pandas as pd

# 원본 데이터셋 로드
file_path = "/content/drive/MyDrive/Data/final_train.csv"
df_original = pd.read_csv(file_path)

# 새로운 데이터셋 생성
test_df = pd.DataFrame()

# 모든 CWE ID 행에 대해 처리
for cwe_id in df_original['CWE ID'].unique():
    # vul이 1인 데이터를 32개 추출
    vul_1_data = df_original[(df_original['CWE ID'] == cwe_id) & (df_original['vul'] == 1)].sample(n=min(32, len(df_original)), random_state=42)
    # vul이 0인 데이터를 16개 추출
    vul_0_data = df_original[(df_original['CWE ID'] == cwe_id) & (df_original['vul'] == 0)].sample(n=min(16, len(df_original)), random_state=42)
    # 추출한 데이터를 새로운 데이터셋에 추가
    test_df = pd.concat([test_df, vul_1_data, vul_0_data])

# 결과를 확인
print(test_df.shape)

# 인덱스 초기화
test_df.reset_index(drop=True, inplace=True)

# 결과를 CSV 파일로 저장
output_file_path = "/content/drive/MyDrive/Data/final_val.csv"
test_df.to_csv(output_file_path, index=False)
```
